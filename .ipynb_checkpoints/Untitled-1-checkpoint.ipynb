```python
import requests
def make_request(page_num):
    query = '''
    query ( $page: Int, $perPage: Int
  , $startDate_greater: FuzzyDateInt
,$countryOfOrigin:CountryCode
,$type:MediaType) {
        Page (page: $page, perPage: $perPage) {
            pageInfo {
                total
                currentPage
                lastPage
                hasNextPage
                perPage
            }
            media (startDate_greater:$startDate_greater,
            countryOfOrigin:$countryOfOrigin,
            type:$type) {
                id
                title {
                    english
                    native
                }
                season
                seasonYear
                type
                format
                status
                episodes
                duration
                chapters
                volumes
                isAdult
                averageScore
                popularity
                source
                countryOfOrigin
                isLicensed
                genres
                startDate {
                    day
                    month
                    year
                  }
                endDate {
                    day
                    month
                    year
                  }
            }
        }
    }
    '''
    variables = {
        "startDate_greater": 20121231 # Fuzzy Date for Dec 31, 2012
  ,"countryOfOrigin": "JP"
  ,"type": "ANIME"
  ,"page": page_num
  ,"perPage": 500
  ,"format": "TV"
    }
    
    url = 'https://graphql.anilist.co'

    response = requests.post(url, json={'query': query, 'variables': variables})
    return response
```

```python
import json
import pandas as pd

df=pd.DataFrame()

for i in range(1,10):
    print(i)
    response=make_request(i)
    json_data = json.loads(response.text)
    df_new=pd.json_normalize(json_data['data']['Page']['media'], meta  =[['english', 'native']])
    df=pd.concat([df,df_new])
df.reset_index(inplace=True,drop=True)
df.sort_values(['startDate.year','startDate.month','startDate.day'])
```

```python
import requests
def make_request(page_num):
    query = '''
    query ($id: Int, $page: Int, $perPage: Int, $search: String) {
        Page (page: $page, perPage: $perPage) {
            pageInfo {
                total
                currentPage
                lastPage
                hasNextPage
                perPage
            }
            media (id: $id, search: $search) {
                id
                title {
                    english
                    native
                }
                season
                seasonYear
                type
                format
                status
                episodes
                duration
                chapters
                volumes
                isAdult
                averageScore
                popularity
                source
                countryOfOrigin
                isLicensed
                genres
                startDate {
                    day
                    month
                    year
                  }
                endDate {
                    day
                    month
                    year
                  }
            }
        }
    }
    '''
    variables = {
        'search': 'my',
        'page': page_num,
        'perPage': 4
    }
    url = 'https://graphql.anilist.co'

    response = requests.post(url, json={'query': query, 'variables': variables})
    return response
```
